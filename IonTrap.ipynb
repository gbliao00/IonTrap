```python
def twofreq(f1,f2):
    import matplotlib.pyplot as plt
    import numpy as np

    x=[i for i in range(3000)]
    y=[np.sin(f1*i)+np.sin(f2*i) for i in range(3000)]
    plt.plot(x,y)
    plt.show()
```

```python
twofreq(0.01,0.001)
```

```python
f1=0.01
f2=0.01

for df in range(0,5):
    df/=100
    twofreq(f1,f2+df)
    
```

```python
import matplotlib.pyplot as plt
import numpy as np

# f1=1;f2=21

def twofreq(f1,f2):
    
    x  = np.linspace(0, 6, 10000) # generates x axis in range 0 to 6 with 20000 points
    y1 = 0.8*np.sin(f1*x) # the first sinus signal with the amplitude 0.8
    y2 = 0.2*np.sin(f2*x) # the second sinus signal with a frequency 20 times higher than the first one and the amplitude of 0.2

    # plot both signals
    plt.plot(x, y1, 'r')
    plt.plot(x, y2, 'b')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('base signals')
    plt.show()


    sum= y1+y2
    plt.plot(x, sum, 'c')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('sum of boath signals')
    plt.show()
```

```python
twofreq(1,31)
```
